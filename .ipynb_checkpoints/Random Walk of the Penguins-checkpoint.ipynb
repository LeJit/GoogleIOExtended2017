# Handling Missing Values

#### Before we can proceed, we need to fill in the missing entries in the time series.

```python
#Taken from the Penguin Benchmark blog post
def preprocess_timeseries(timeseries, first_year, fillna_value = 0.0):
    
    timeseries.columns = timeseries.columns.astype(int)
    
    timeseries = timeseries.loc[:, timeseries.columns >= first_year]
    
    timeseries.fillna(method = "ffill", axis = 1, inplace = True)
    timeseries.fillna(method = "bfill", axis = 1, inplace = True)
    
    timeseries.fillna(fillna_value, inplace = True)
    
    return timeseries
```

```python
nest_counts = preprocess_timeseries(nest_counts, 1985)
e_n_values = preprocess_timeseries(e_n_values, 1985)
```

# The Metric: Adjusted MAPE

For this challenge, we are going to use the Adjusted Mean Absolute Percentage Error (Adjusted MAPE) to evaluate the performance of our models.

\begin{eqnarray}
\text{Adjusted MAPE} = \frac{1}{N} \sum_{i=1} ^N [\frac{\frac{|\hat{y}_i - y_i|}{max(1,y_i)}}{e_i}]
\end{eqnarray}

The Adjusted MAPE calculation can be broken down into three steps:
1. Compute the absolute difference between our predicted value and the true value.
2. From the Absolute Error, compute the percent error (divide by true value; Account for division by zero)
3. Adjust the Percent Error based on the accuracy associated with the true value.

```python
#This code comes from the Penguin Benchmark Blog post
def adjusted_MAPE(y_true, y_pred, errors):
    #Mask for indices of non-NAN values
    not_nan_mask = ~np.isnan(y_true)
    #Calculate the Absolute Error between the true values and the predicted values
    abs_error = np.abs(y_true[not_nan_mask] - y_pred[not_nan_mask])
    #Calculate the percent error
    pct_error = abs_error / np.maximum(1, y_true[not_nan_mask])
    #Adjust the percenter error by the accuracy
    adj_error = pct_error / errors[not_nan_mask]
    #Return the average adjusted error
    return np.mean(adj_error)
```

# ARIMA

Autoregressive Integrated Moving Averages (ARIMA) is a commonly used time series analysis method. There are three mains parts to ARIMA.

1. AR - "Autoregressive" refers to previous time steps we look back.
2. I - "Integrated" means the data has been replaced with difference between it and the previous values.
3. MA - "Moving Averages" refers to the number of historical error terms we use.

```python
#Convert Pandas Series into an unstacked DataFrame; To work with PyFlux
acun_adelie = nest_counts.iloc[0]
acun_adelie_df = acun_adelie.to_frame()
acun_adelie_df.unstack(level = -1)
```

Let's construct an ARIMA(3,0,3) model. This means we will look at the three previous timesteps when forecasting a future year.

```python
model = pf.ARIMA(data = acun_adelie_df, ar = 3, ma = 3, integ = 0)
```

After producing our model, we can estimate the best parameters and see the results.

```python
x = model.fit("MLE")
x.summary()
```

# Antartica's Next Top Model

Now, let's train an ARIMA model on each of our Time Series. Since our model has three parameters, the training step would normally involve finding the optimal combination of these three parameters. For now, we will focus on optimizing the Autoregressive parameter (AR).

We will break the training step into four steps:

1. Take the time series, and split into two smaller time series (before/after 2005)
2. Build an ARIMA model of the form ARIMA(*p*,0,3), and train it on the before 2005 time series.
3. Score the model's performance on the After 2005 time series using the Adjusted MAPE.
4. Build an ARIMA model with the best AR parameter and train it on all the data (through 2013)
4. Return the model along with the site information.

```python
def train_model_ARIMA(ts, accuracy, split_year = 2005):
    #Split before 2005
    train_ts = ts.iloc[ts.index < split_year]
    #Convert the Before 2005 TS into a DataFrame
    train_df = train_ts.to_frame()
    train_df.unstack(level = -1)
    #Split after 2005
    cv_ts = ts.iloc[ts.index >= split_year]
    cv_acc = accuracy.iloc[accuracy.index >= split_year]
    #Store the Adjusted MAPE scores
    amapes_ = []
    #Iterate through 5 different values for p
    for ar in range(5):
        model = pf.ARIMA(data = train_df, ar = ar, ma = 3, integ = 0)
        model.fit("MLE")
        #Predict on the future years (2005- 2013)
        y_pred = model.predict(h = 9).values.flatten()
        #Score the model using Adjusted MAPE
        amapes_.append(adjusted_MAPE(cv_ts.values, y_pred, cv_acc.values))
    #Find the best value of p for the Autoregressive component
    best_ar = np.argmin(amapes_)
    #Build the optimal model
    ts_df = ts.to_frame()
    ts_df.unstack(level = -1)
    
    model = pf.ARIMA(data = ts_df, ar = best_ar, ma = 3, integ = 0)
    return model
```

Now, let's build all the models.

```python
models = {}

for i, row in tqdm_notebook(nest_counts.iterrows(), total = nest_counts.shape[0]):
    accuracy = e_n_values.loc[i]
    models[i] = train_model_ARIMA(row, accuracy)
```

As you can see, it took sixteen minutes to train all 648 ARIMA models on my local machine.

# Make a submission

Now that we have trained our models, we need to make a submission file.

```python
submission_format = pd.read_csv(os.path.join('data','submission_format.csv'),index_col=[0, 1])
```

```python
submission_format.head()
```

### For each model, we are going to forecast four years into the future, and store those results.

```python
predictions = []

for i, row in tqdm_notebook(submission_format.iterrows(),
                            total=submission_format.shape[0]):
    # get the model for this site + common_name
    model = models[i]
    model.fit("MLE")
    
    # make predictions using the model
    row_predictions = model.predict(h = 4).values.flatten()
    # keep our predictions, rounded to nearest whole number
    predictions.append(np.round(row_predictions))
```

Create a DataFrame to store all our predictions. This will make it easier to format the CSV file.

```python
prediction_df = pd.DataFrame(predictions,
                             index=submission_format.index,
                             columns=submission_format.columns)
```

```python
prediction_df.head()
```

Save the results to a CSV file and upload it to DrivenData's submission server.

```python
prediction_df.to_csv("output/predictions.csv")
```
